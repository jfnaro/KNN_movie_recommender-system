```python
movies_key = pd.read_csv('ml-100k/u.item', sep='|', names=m_cols, encoding='latin-1', usecols=['movieId', 'title'])
movies_key
```

```python
users_key= pd.read_csv('ml-100k/u.data', delimiter='\t', names=['userId', 'movieId', 'rating', 'timestamp'], usecols=['userId'])
users_key.sort_values('userId', inplace=True)
users_key.drop_duplicates(inplace=True)
users_key.reset_index(drop=True, inplace=True)
users_key
```

```python
ratings_df = pd.read_csv('ml-100k/u.data', delimiter='\t', names=['userId', 'movieId', 'rating', 'timestamp'])
ratings_df.drop('timestamp', axis=1, inplace=True)
ratings_df.sort_values('userId', inplace=True)
ratings_df.reset_index(drop=True, inplace=True)
ratings_df
```

```python
chosen_user = 3
```

```python
user_profiles = ratings_df.pivot(index='movieId', columns='userId', values='rating').fillna(0)
user_profiles = user_profiles.transpose()
user_profiles
```

### The description of the data below is largely useless, because the zeros representing unwatched videos are mixed in with the actual reviews.

```python
user_profiles.describe()
```

# Parts 1.2 and 1.5 are integrated into Parts 1.3 and 1.4 below for more convenient reading.

### For this part, we will be finding the nearest and the farthest users.

```python
user_neighbors = NearestNeighbors()
user_neighbors.fit(user_profiles)
```

## Because the possible values users can have for each movie are the integers from 0 to 5 with 0 being unseen movies and 1 being the lowest rating a user can give a movie, we can reasonably assume that the farther a user's profile is from a completely blank profile, the more they are using the service and the happier they are with the movies they watch. These seem like our ideal satisfied customers, and we want new people to be this satisfied as well. As such, we will go through the profiles of the 5 users who are the furthest from the origin and recommend two of each of their top rated movies.

```python
user_ratings = ratings_df.loc[ratings_df['userId'] == chosen_user]
user_ratings.sort_values('rating', inplace=True, ascending=False)
user_ratings.reset_index(drop=True, inplace=True)
```

```python
similar_users = user_neighbors.kneighbors([np.zeros(len(user_profiles.columns))], len(user_profiles) - 5, return_distance=False)

optimal_users = [user for user in users_key.index if user not in similar_users]
optimal_users

real_users = []
for user in optimal_users:
    real_users.append(users_key.loc[user, 'userId'])
recs = []
for user in real_users:
    user_id = users_key.loc[user, 'userId']
    
    close_user_ratings = ratings_df.loc[ratings_df['userId'] == user]
    close_user_ratings.sort_values('rating', inplace=True, ascending=False)
    close_user_ratings.reset_index(drop=True, inplace=True)
    
    index = 0
    user_favs = 0
    while index < len(close_user_ratings) and user_favs < 2 and close_user_ratings.loc[index, 'rating'] > 3:
        if close_user_ratings.loc[index, 'movieId'] not in recs and close_user_ratings.loc[index, 'movieId'] not in user_ratings['movieId']:
            recs.append(close_user_ratings.loc[index, 'movieId'])
            user_favs += 1
        index += 1
            
recs
```

```python
for movie in recs:
    print(movies_key.loc[movie, 'title'])
```

### Here we have it, a list of 10 favorites amongst 5 of the most active users. This should be a good place for new users to start. It's also easy to find more recommendations simply by increasing the amount of top users to draw from or by increasing the amount of recommendations to get from each top user.

## Collaborative

### It is reasonable to assume that similar users will give the same movies similar ratings. Therefore, if we find which users are the most like the user in question, we can find the movies they rated the highest and recommend them to the user with the understanding that the user is likely to give those movies the same ratings.

```python
user_profiles.loc[chosen_user]
```

### For this, we're just using the 3 most similar users to get 3 recommendations each. We could easily be expanded to more users, which would increase the amount of recommendations we could give.

```python
similar_users = user_neighbors.kneighbors([user_profiles.loc[chosen_user]], 4, return_distance=False)
real_users = []
for user in similar_users[0]:
    real_users.append(users_key.loc[user, 'userId'])
real_users.remove(chosen_user)
recs = []
for user in real_users:
    user_id = users_key.loc[user, 'userId']
    
    close_user_ratings = ratings_df.loc[ratings_df['userId'] == user]
    close_user_ratings.sort_values('rating', inplace=True, ascending=False)
    close_user_ratings.reset_index(drop=True, inplace=True)
    close_user_ratings
    
    index = 0
    user_favs = 0
    while index < len(close_user_ratings) and user_favs < 3 and close_user_ratings.loc[index, 'rating'] > 3:
        if close_user_ratings.loc[index, 'movieId'] not in recs and close_user_ratings.loc[index, 'movieId'] not in user_ratings['movieId']:
            recs.append(close_user_ratings.loc[index, 'movieId'])
            user_favs += 1
        index += 1
            
recs
```

```python
for movie in recs:
    print(movies_key.loc[movie, 'title'])
```

## Content-based

### The idea here is that if a user rated a movie highly, then the user can be expected to give a similar rating to similar movies. As such, we recommend similar movies.

### We are now looking at movie similarities.

```python
movie_neighbors = NearestNeighbors()
movie_neighbors.fit(movies_df)
```

### We start by taking 3 of the users top rated movies. For each movie, we then find the 3 most similar unrecommended movies.

```python
recs = []
for row in range(3):
    print(movies_key.loc[movies_key['movieId'] == user_ratings.loc[row, 'movieId'], 'title'])
    matches = np.squeeze(movie_neighbors.kneighbors([movies_df.loc[user_ratings.loc[row, 'movieId']]], 5+(5*row), return_distance=False))
    count = 0
    i = 0
    while count < 3 and i < len(matches):
        if matches[i] not in recs and movies_key.loc[matches[i], 'movieId'] not in user_ratings['movieId'].to_numpy():
            recs.append(matches[i])
            count += 1
        i += 1
recs
```

### Below are our recommendations. Some of the similarities are harder to see, depending on how familiar you are with the movies, but the movies have clear similarities to the ones the recommendations were based on (listed above).

```python
for movie in recs:
    print(movies_key.loc[movie, 'title'])
```
